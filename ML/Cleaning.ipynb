```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

# Use 'ggplot' style for plots
mpl.style.use('ggplot')

# Load dataset
car = pd.read_csv('quikr_car.csv')

# Display first few rows
car.head()


# Check shape of the dataset
car.shape  # (892, 6)

# Check data info
car.info()

# Keep only rows where 'year' is numeric
car = car[car['year'].str.isnumeric()]

# Convert 'year' to integers
car['year'] = car['year'].astype(int)

# Remove rows where price is 'Ask For Price'
car = car[car['Price'] != 'Ask For Price']

# Remove commas and convert to integer
car['Price'] = car['Price'].str.replace(',', '').astype(int)

# Keep only numeric values in 'kms_driven'
car['kms_driven'] = car['kms_driven'].str.split().str.get(0).str.replace(',', '')

# Remove rows with non-numeric 'kms_driven' values
car = car[car['kms_driven'].str.isnumeric()]

# Convert to integer
car['kms_driven'] = car['kms_driven'].astype(int)

# Remove rows with missing fuel_type
car = car[~car['fuel_type'].isna()]

car.shape  # Should be (816, 6)

car['name'] = car['name'].str.split().str.slice(start=0, stop=3).str.join(' ')

# Reset index after cleaning
car = car.reset_index(drop=True)
```
